```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
import neuralnetwork
```

## This part draws a circle. 
Conveniently, it will increase the thickness as the circle grows larger. This was my attempt to try use a circle, but I ultimately discarded it.

```python
IMG_LEN = 28

blankimage = np.zeros(shape=[IMG_LEN, IMG_LEN])

x_centre = random.randint(10, 17)
y_centre = random.randint(10, 17)


rad_max = max(x_centre, y_centre)
if (rad_max > 12):
    rad = random.randint(5, (IMG_LEN - 3) - rad_max)
else:
    rad = random.randint(5, rad_max)

if rad > 5:
    thick = 2
else:
    thick = 1


cv2.circle(blankimage, center=(x_centre, y_centre), radius=rad, color=(255,255,255), thickness=thick)

circle = blankimage

plt.imshow(circle)
plt.show()
```

## This code will draw a square. 
Thickness of 2 was useless but thickness of 1 was really useful 

```python
blankimage = np.zeros(shape=[28, 28])

isbig = random.randint(0, 1)

if isbig:
    sq_size = 12
    x_pt1 = random.randint(1,15)
    y_pt1 = random.randint(1,15)   
else:
    sq_size = 7
    x_pt1 = random.randint(1,20)
    y_pt1 = random.randint(1,20)   

x_pt2 = x_pt1 + sq_size
y_pt2 = y_pt1 + sq_size

cv2.rectangle(blankimage, pt1=(x_pt1, y_pt1), pt2=(x_pt2, y_pt2), color=(255, 255, 255), thickness=2)

square = blankimage

plt.imshow(square)
plt.show()
```

## The following 4 cells shows each stimuli per dataset

### Train (straight line)

```python
test_data_file = open("datasets/train_str8_10.csv", 'r')
test_data_list = test_data_file.readlines()
test_data_file.close() 

count = 0
labels = [0, 1, 2, 3]

for record in test_data_list:
    all_values = record.split(',')
    correct_label = int(float(all_values[0]))
    
    if not labels:
        break

    elif correct_label in labels:

        inputs = np.asfarray(all_values)    
        inputs = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
        
        image = np.asfarray(inputs.reshape((28,28)))
        plt.imshow(image)
        plt.figure(count)
        
        count = count + 1
        labels.remove(correct_label)
        
        print("    Label = ",correct_label)
        plt.show()
```
